**Question 2:** In how many different states do the customers live in?

```python
# open file and init vars
import csv

datafile = "./sample_data/customers.csv"

unique_states = set()

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    next(csvreader)

    for row in csvreader:
        if not row: continue
        state = row[4]

        state_cleaned = state.strip()

        if state.strip() != "":
            unique_states.add(state_cleaned)

print(f"Question 2 (Number if unique states): {len(unique_states)}")

# evaluate the contents of the file
# 16 states
```

**Question 3** What is the state with most customers?

```python
# open file and init vars
import csv

datafile = "./sample_data/customers.csv"

state_counts = {}

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)

  for row in csvreader:
    state = row[4]
    if state in state_counts:
      state_counts[state] += 1
    else:
      state_counts[state] = 1
state_with_most_customers = max(state_counts, key=state_counts.get)

print(f"Question 3 (State with most customers): {state_with_most_customers}")

# evaluate the contents of the file
# Question 3 (State with most customers): CA
```

**Question 4** What is the state with the least customers?

```python
# open file and init vars
import csv

datafile = "./sample_data/customers.csv"

state_counts = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    next(csvreader)

    for row in csvreader:
        if not row: continue

        state = row[4].strip()
        if state == "": continue

        if state in state_counts:
            state_counts[state] += 1
        else:
            state_counts[state] = 1

min_customers = min(state_counts.values())

tied_states = []
for state, count in state_counts.items():
    if count == min_customers:
        tied_states.append(state)

print(f"Minimum customer count is: {min_customers}")
print(f"States with the least customers: {tied_states}")

# evaluate the contents of the file
# Question 4 States with the least customers: ['WA', 'NH', 'ID', 'OR', 'MA', 'IN']
```

**Question 5:** What is the most common last name?

```python
# open file and init vars

import csv

datafile = "./sample_data/customers.csv"

last_name_counts = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    next(csvreader)

    for row in csvreader:
        last_name = row[2]

        if last_name in last_name_counts:
            last_name_counts[last_name] += 1
        else:
            last_name_counts[last_name] = 1

most_common_last_name = max(last_name_counts, key=last_name_counts.get)

print(f"Question 5 (Most common last name): {most_common_last_name}")

# evaluate the contents of the file
# Question 5 (Most common last name): Smith
```

# Exercise 2: Processing Orders data (difficulty high)

The second sample files contains orders placed by customers from the first file. Be careful, this file has many rows and you most likely should not print the contents of the whole file.

**Question #1:** How many unique orders are in the orders.csv file?

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

**Question #3:** What is the highest number of items per order?

**Question #4:** What is the number of orders placed in October 2021?

**Question #5:** Which customer spent the most amount of money in 2021?

**Question #6:** Historically, what is the best month for sales?

Once you get your answers, remember to go back to the course and introduce them in the multiple choice quiz

```python
import csv

datafile = "./sample_data/orders.csv"

items_per_order = {}
orders_oct_2021 = set()
customer_spending_2021 = {}
sales_per_month = {}

with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    next(csvreader)

    for row in csvreader:
        if not row: continue

        customer_id = row[0]
        order_id = row[1]
        date = row[2]
        price = float(row[5])

        if order_id in items_per_order:
            items_per_order[order_id] += 1
        else:
            items_per_order[order_id] = 1

        if len(date) >= 7:

            if date.startswith("2021-10"):
                orders_oct_2021.add(order_id)

            if date.startswith("2021"):
                if customer_id in customer_spending_2021:
                    customer_spending_2021[customer_id] += price
                else:
                    customer_spending_2021[customer_id] = price

            month = date[:7]
            if month in sales_per_month:
                sales_per_month[month] += price
            else:
                sales_per_month[month] = price

# Q1
print(f"Question 1 (Unique orders): {len(items_per_order)}")

# Q2
total_items = sum(items_per_order.values())
average = total_items / len(items_per_order)
print(f"Question 2 (Average items per order): {round(average, 2)}")

# Q3
print(f"Question 3 (Highest number of items per order): {max(items_per_order.values())}")

# Q4
print(f"Question 4 (Orders in October 2021): {len(orders_oct_2021)}")

# Q5
best_customer = max(customer_spending_2021, key=customer_spending_2021.get)
mount_spending = customer_spending_2021[best_customer]
print(f"Question 5 (Customer with most spending in 2021): {best_customer}")
print(f"Question 5 (Spending in 2021): {mount_spending}")

# Q6
best_month = max(sales_per_month, key=sales_per_month.get)
print(f"Question 6 (Best month for sales): {best_month}")

# Question 1 (Unique orders): 16672
# Question 2 (Average items per order): 1.76
# Question 3 (Highest number of items per order): 35
# Question 4 (Orders in October 2021): 267
# Question 5 (Customer with most spending in 2021): 5172443
# Question 6 (Best month for sales): 2022-01


```
